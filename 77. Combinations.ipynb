```python
n = 4
k = 3
res = []

def backtrack(start, comb):
    print("start : {}".format(start))
    print("comb : {}".format(comb))
    if len(comb)==k:
        print("comb : inside if {}".format(comb))
        print("res : inside if {}".format(res))
        res.append(comb.copy())
        print("res : inside if after append {}".format(res))
        return
    for i in range(start, n+1):
        print("i: {}".format(i))
        comb.append(i)
        print("comb : inside for {}".format(comb))
        print("i+1 : {}".format(i+1))
        backtrack(i+1, comb)
        print("below backtrack")
        comb.pop()
        print("comb : below backtrack for {}".format(comb))
        print("i : {}".format(i))

backtrack(1, [])
print(res)
```
